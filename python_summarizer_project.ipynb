```python
!pip install kagglehub datasets transformers

import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("gowrishankarp/newspaper-text-summarization-cnn-dailymail")

print("Path to dataset files:", path)

# List files inside the dataset folder
for root, dirs, files in os.walk(path):
    for file in files:
        print(os.path.join(root, file))

# Now directly use the 'path' you downloaded
# If "validation.csv" is inside the 'cnn_dailymail' folder, then do:

validation_csv_path = os.path.join(path, "cnn_dailymail", "validation.csv")
df = pd.read_csv(validation_csv_path)

df.head()
```

```python
df = df.dropna()
df = df[['article', 'highlights']]  # only keep relevant columns
```

```python
from transformers import pipeline

# Load a summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Try it on one article
article = df['article'][0]
summary = summarizer(article, max_length=130, min_length=30, do_sample=False)

print("Original:\n", article)
print("\nSummary:\n", summary[0]['summary_text'])
```

```python
from transformers import pipeline

# Load pre-trained summarizer
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def summarize_text(input_text):
    # Truncate if input is too long for the model
    input_text = input_text[:2000]
    summary = summarizer(input_text, max_length=150, min_length=40, do_sample=False)
    return summary[0]['summary_text']
```

```python
import gradio as gr

interface = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(lines=15, placeholder="Paste your article or text here..."),
    outputs=gr.Textbox(label="Summary"),
    title="📰 Text Summarizer",
    description="Paste a news article, blog post, or any long text to get a summary!"
)

interface.launch()
```
